```python
df = pd.read_csv('/kaggle/input/youtube-video-and-channel-analytics/YouTubeDataset_withChannelElapsed.csv')
df.info()
```

```python
df = df.drop_duplicates(subset=['videoId'])
df[['videoLikeCount','videoDislikeCount','VideoCommentCount','videoViewCount']]=df[['videoLikeCount','videoDislikeCount','VideoCommentCount','videoViewCount']].replace(-1, None)
```

```python
ret1 = df['videoLikeCount'].mean()
ret1
```

```python
engagements = df['videoLikeCount'].sum()+df['videoDislikeCount'].sum()+df['VideoCommentCount'].sum()
views = df['videoViewCount'].sum()
ret2 = engagements/views
ret2 = round(ret2,4)
ret2
```

```python
K = 2**10
M = 2**20
ret3 = round((500 + 200*K) / M,4)
ret3
```

```python

ret4 = engagements*ret3
ret4
```

```python
span = 5*365*24*60
ret5 = df['videoViewCount'].sum()/span
ret5
```

```python
ret6 = ret4/span
ret6
```
